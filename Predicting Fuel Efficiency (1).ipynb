### Feature Engineering on model year feature.

```python
def model_year(year):
    return 1900 + int(year)

data["model year"] =  data["model year"].apply(model_year)
data["model year"] = pd.to_numeric(pd.to_datetime(data["model year"]))
```

```python
data.drop("car name",axis=1,inplace=True)
```

### Making model fit for training.

```python
data.head()
```
